```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('D:/Programming/Jupyter/Hello/data/covid_19_data.csv')
```

```python
df.head(50)
```

```python
df.drop(['Last Update'], axis=1, inplace=True)
```

```python
df.head(50)
```

```python
df.rename(columns={'Province/State': 'State', 'Country/Region': 'Country'}, inplace=True)
```

```python
df.head(50)
```

```python
imputer = SimpleImputer(strategy='constant')
```

```python
df2 = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

```python
df3 = df2.groupby(['Country'])[['Country', 'Confirmed', 'Deaths', 'Recovered']].sum()
```

```python
df3.head(50)
```

```python
countries = df3['Country'].unique()
len(countries)
```

```python
for idx in range(0, len(countries)):
#     C = df3[df3['Country']==countries[idx]].reset_index()
# This is not going to show the correct graphics 
# because I did not find the data used in the example
    C = df3
    plt.scatter(np.arange(0, len(C)), C['Confirmed'], color='blue', label='Confirmed')
    plt.scatter(np.arange(0, len(C)), C['Recovered'], color='green', label='Recovered')
    plt.scatter(np.arange(0, len(C)), C['Deaths'], color='red', label='Deaths')
    plt.title(countries[idx])
    plt.xlabel('Days since the first suspect')
    plt.ylabel('Number of cases')
    plt.legend()
    plt.show()
```
